##  Analyze historical sales data to forecast future demand for different products accurately

```python
import pandas as pd

# Load sales data
#sales_data = pd.read_csv("/mnt/data/Salesfinalcopy.csv")

# Prepare date and product info
sales_data['SalesDate'] = pd.to_datetime(sales_data['SalesDate'])
sales_data['YearMonth'] = sales_data['SalesDate'].dt.to_period('M').astype(str)
sales_data['Product'] = sales_data['Brand'].astype(str) + " - " + sales_data['Description']

# Aggregate monthly sales
monthly_sales = sales_data.groupby(['YearMonth', 'Product'])['SalesQuantity'].sum().reset_index()
monthly_sales['YearMonth'] = pd.to_datetime(monthly_sales['YearMonth'])

# Select top 3 products by total sales
top_products = (
    monthly_sales.groupby('Product')['SalesQuantity']
    .sum()
    .sort_values(ascending=False)
    .head(3)
    .index
)

# Forecast for 6 months using historical average
forecast_horizon = 6
forecast_rows = []

for product in top_products:
    product_sales = monthly_sales[monthly_sales['Product'] == product]
    avg_monthly_sales = product_sales['SalesQuantity'].mean()

    for i in range(1, forecast_horizon + 1):
        forecast_rows.append({
            'Product': product,
            'ForecastMonth': pd.Timestamp('2017-01-01') + pd.DateOffset(months=i-1),
            'ForecastQuantity': round(avg_monthly_sales, 2)
        })

# Create DataFrame
forecast_df = pd.DataFrame(forecast_rows)

print(forecast_df.head(6))
```

###   Demand Patterns and Forecasting

Key Insights:
1. Products with consistent past sales can be reliably forecasted using historical averages.

2. Forecasted demand helps align purchasing plans with expected needs, avoiding under- or overstocking.

3. Short time series data limits seasonal modeling, but monthly average-based forecasts work well for stable SKUs.

Recommendations:
1. Use rolling 3- to 6-month averages to plan upcoming purchases.

2. Forecasts should inform reorder points and EOQ timing.

3. Add seasonality adjustments if more data becomes available.

## ABC Analysis

```python


#  Create a unique product identifier
sales_data['Product'] = sales_data['Brand'].astype(str) + " - " + sales_data['Description']

#  Calculate total sales value per product
product_sales = sales_data.groupby('Product')['SalesDollars'].sum().reset_index()
product_sales = product_sales.sort_values(by='SalesDollars', ascending=False).reset_index(drop=True)

#  Calculate cumulative sales and percentages
product_sales['CumulativeSales'] = product_sales['SalesDollars'].cumsum()
total_sales = product_sales['SalesDollars'].sum()
product_sales['CumulativePercent'] = 100 * product_sales['CumulativeSales'] / total_sales

#  Define ABC classification based on cumulative percentage
def classify_abc(percent):
    if percent <= 80:
        return 'A'
    elif percent <= 95:
        return 'B'
    else:
        return 'C'

product_sales['ABC_Category'] = product_sales['CumulativePercent'].apply(classify_abc)

# Output result
print(product_sales[['Product', 'SalesDollars', 'CumulativePercent', 'ABC_Category']].head(10))
```

## Visualize the ABC analysis

```python

plt.figure(figsize=(14, 6))
bars = plt.bar(product_sales.index, product_sales['SalesDollars'], color='skyblue')

# Color bars by category
for i, category in enumerate(product_sales['ABC_Category']):
    if category == 'A':
        bars[i].set_color('green')
    elif category == 'B':
        bars[i].set_color('orange')
    else:
        bars[i].set_color('red')

# Add threshold lines
plt.axvline(x=len(product_sales[product_sales['ABC_Category'] == 'A']), color='gray', linestyle='--', label='A/B Threshold')
plt.axvline(x=len(product_sales[product_sales['ABC_Category'].isin(['A', 'B'])]), color='red', linestyle='--', label='B/C Threshold')

# Labels and formatting
plt.title('ABC Classification - Bar Chart')
plt.xlabel('Products (sorted by Sales Value)')
plt.ylabel('Sales Dollars')
plt.legend()
plt.tight_layout()
plt.show()
```

## Determine the optimal order quantity for raw materials to minimize ordering and carrying costs.

```python

sales_data['Product'] = sales_data['Brand'].astype(str) + " - " + sales_data['Description']

# Calculate annual demand (units sold in a year per product)
annual_demand = sales_data.groupby('Product')['SalesQuantity'].sum().reset_index()
annual_demand.columns = ['Product', 'AnnualDemand']


purchase_prices['Product'] = purchase_prices['Brand'].astype(str) + " - " + purchase_prices['Description']

# Merge sales demand with purchase prices
eoq_data = pd.merge(annual_demand, purchase_prices[['Product', 'PurchasePrice']], on='Product', how='inner')

# Assumptions:
ordering_cost = 100             # Cost per order (can be customized)
carrying_cost_per_unit = 5      # Annual holding cost per unit (estimated)

# EOQ formula
eoq_data['EOQ'] = np.sqrt((2 * ordering_cost * eoq_data['AnnualDemand']) / carrying_cost_per_unit)

# Round EOQ values for clarity
eoq_data['EOQ'] = eoq_data['EOQ'].round(2)

# Show final table
print(eoq_data[['Product', 'AnnualDemand', 'PurchasePrice', 'EOQ']].head(10))
```

2. This reflects moderate annual demand and a high carrying cost assumption ($5/unit/year), indicating that overstocking would be costly.

3. Products with higher annual demand (e.g., F Coppola Dmd Ivry Cab Svgn) have slightly higher EOQs, aligning with inventory theory.

Recommendations:

3 For Category A items (from ABC), place smaller, more frequent orders to maintain high service levels with reduced carrying costs.

## Calculate the reorder point for each product to avoid stockouts (for the top-selling products, using an assumed lead time of 15 days and rounds the final ROP to whole numbers for practical use).

```python

import numpy as np
# Create product identifier
sales_data['Product'] = sales_data['Brand'].astype(str) + " - " + sales_data['Description']

# Aggregate annual demand
top_sellers = sales_data.groupby('Product')['SalesQuantity'].sum().reset_index()
top_sellers.columns = ['Product', 'AnnualDemand']

# Select top 10 products
top_sellers = top_sellers.sort_values(by='AnnualDemand', ascending=False).head(10)

# Step 1: Assumed lead time (in days)
top_sellers['AvgLeadTimeDays'] = 15

# Step 2: Daily demand
top_sellers['DailyDemand'] = top_sellers['AnnualDemand'] / 365

# Step 3: Calculate ROP and safety stock (20% buffer)
top_sellers['ROP_WithoutSafety'] = top_sellers['DailyDemand'] * top_sellers['AvgLeadTimeDays']
top_sellers['SafetyStock'] = 0.2 * top_sellers['ROP_WithoutSafety']
top_sellers['ReorderPoint'] = (top_sellers['ROP_WithoutSafety'] + top_sellers['SafetyStock']).round()

# Final output (rounded ROPs for easier use in ordering)
top_sellers = top_sellers[['Product', 'AnnualDemand', 'AvgLeadTimeDays', 'DailyDemand', 'SafetyStock', 'ReorderPoint']]

# Show result
print(top_sellers.head())

# Calculate EOQ (Economic Order Quantity)
ordering_cost = 100  # Assumed ordering cost per order
carrying_cost_per_unit = 5  # Assumed annual carrying cost per unit
top_sellers['EOQ'] = np.sqrt((2 * ordering_cost * top_sellers['AnnualDemand']) / carrying_cost_per_unit)

# Plot EOQ and ROP side by side
plt.figure(figsize=(14, 6))
x = range(len(top_sellers))
bar_width = 0.35

# Plot bars
plt.bar([p - bar_width/2 for p in x], top_sellers['ReorderPoint'], width=bar_width, label='Reorder Point', color='skyblue')
plt.bar([p + bar_width/2 for p in x], top_sellers['EOQ'], width=bar_width, label='EOQ', color='orange')

# Customize chart
plt.title('Reorder Point vs EOQ for Top 10 Selling Products')
plt.xlabel('Product')
plt.ylabel('Units')
plt.xticks(ticks=x, labels=top_sellers['Product'], rotation=45, ha='right')
plt.legend()
plt.grid(True, axis='y')
plt.tight_layout()

# Show plot
plt.show()
```

###  Business Insights

1. These products have low daily demand, so their ROP values are also low (typically <1 unit).

2. ROP < 1 unit implies they should be reordered only when stock is very low, but in larger quantities (based on EOQ).

3. For practical purposes, consider setting a minimum ROP = 1 for any product that sells regularly.

 ## Assess the lead time for raw materials and production to optimize inventory levels

```python


# Ensure dates are parsed correctly
purchase_final['PODate'] = pd.to_datetime(purchase_final['PODate'], dayfirst=True, errors='coerce')
purchase_final['InvoiceDate'] = pd.to_datetime(purchase_final['InvoiceDate'], dayfirst=True, errors='coerce')

# Calculate lead time
purchase_final['LeadTimeDays'] = (purchase_final['InvoiceDate'] - purchase_final['PODate']).dt.days

# Filter only valid lead times
valid_lead_times = purchase_final[(purchase_final['LeadTimeDays'] > 0) & (purchase_final['LeadTimeDays'] <= 365)].copy()

# Create Product Name
valid_lead_times['Product'] = valid_lead_times['Brand'].astype(str) + " - " + valid_lead_times['Description']

# Group by Product and Vendor
lead_time_analysis = valid_lead_times.groupby(['VendorName', 'Product']).agg(
    AvgLeadTime=('LeadTimeDays', 'mean'),
    MinLeadTime=('LeadTimeDays', 'min'),
    MaxLeadTime=('LeadTimeDays', 'max'),
    StdLeadTime=('LeadTimeDays', 'std'),
    OrderCount=('LeadTimeDays', 'count')
).reset_index()

# Round for better readability
lead_time_analysis[['AvgLeadTime', 'MinLeadTime', 'MaxLeadTime', 'StdLeadTime']] = \
    lead_time_analysis[['AvgLeadTime', 'MinLeadTime', 'MaxLeadTime', 'StdLeadTime']].round(2)

print(lead_time_analysis.head(10))  # Display top 10 lead times by vendor and product
```

###  Business Insights from Lead Time Assessment

 1. Long Average Lead Times Indicate Procurement Bottlenecks
Products with high AvgLeadTime (e.g., 30+ days) may:

 Be sourced internationally or from slow suppliers

 Require longer planning cycles and more advance ordering

 Recommendation:

Increase safety stock for these items

Initiate earlier reordering or explore alternative local suppliers

 2. High Lead Time Variability = High Supply Chain Risk
A high StdLeadTime means inconsistency in delivery performance:

Vendors may not meet deadlines

Delays can cause stockouts even when you reorder on time

 Recommendation:

For such products, increase buffer stock (safety stock)

Negotiate SLAs (Service Level Agreements) with vendors to improve consistency

 3. Low and Stable Lead Times Reduce the Need for Buffer Stock
Products with:

Low AvgLeadTime (e.g., 510 days)

Low StdLeadTime (e.g., < 2 days)

are ideal for lean inventory practices

 Recommendation:

Reduce safety stock for these items

Reorder more frequently with confidence (lower ROPs)

 4. Vendors with High Max Lead Time Should Be Monitored Closely
If MaxLeadTime is much higher than average, it could mean:

Occasional severe delays

Poor communication or logistics issues

 Recommendation:

Flag these vendors for review

Diversify suppliers for critical items

Use max lead time for worst-case ROP calculations

 ## Calculate inventory turnover ratios to assess the efficiency of inventory management

```python
import pandas as pd

# Load your datasets
begin_inventory = pd.read_csv(r'C:\Users\GHANESH\Downloads\BegInvFINAL12312016.csv')
end_inventory = pd.read_csv(r'C:\Users\GHANESH\Downloads\EndInvFINAL12312016.csv')
invoice_purchases = pd.read_csv(r'C:\Users\GHANESH\Downloads\InvoicePurchases12312016.csv')

# Step 1: Calculate inventory values
begin_inventory['InventoryValue'] = begin_inventory['onHand'] * begin_inventory['Price']
end_inventory['InventoryValue'] = end_inventory['onHand'] * end_inventory['Price']

# Step 2: Compute total beginning and ending inventory values
total_beg_inventory = begin_inventory['InventoryValue'].sum()
total_end_inventory = end_inventory['InventoryValue'].sum()

# Step 3: Average inventory value
average_inventory = (total_beg_inventory + total_end_inventory) / 2

# Step 4: Estimate COGS using total invoice purchases
cogs = invoice_purchases['Dollars'].sum()

# Step 5: Calculate inventory turnover ratio
inventory_turnover_ratio = cogs / average_inventory

# Step 6: Calculate days in inventory
days_inventory = 365 / inventory_turnover_ratio

# Step 7: Create output summary
turnover_summary = {
    "Beginning Inventory Value ($)": round(total_beg_inventory, 2),
    "Ending Inventory Value ($)": round(total_end_inventory, 2),
    "Average Inventory Value ($)": round(average_inventory, 2),
    "Estimated COGS ($)": round(cogs, 2),
    "Inventory Turnover Ratio": round(inventory_turnover_ratio, 2),
    "Days in Inventory": round(days_inventory, 2)
}

# Print summary
for key, value in turnover_summary.items():
    print(f"{key}: {value}")
```
